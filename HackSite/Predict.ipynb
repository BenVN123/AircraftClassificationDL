```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten, Conv3D, Conv2D, MaxPooling2D, AveragePooling2D
from keras.optimizers import Adam
import os
import numpy
import pickle
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
X = pickle.load(open('X.pickle', 'rb'))
y = pickle.load(open('y.pickle', 'rb'))

X = np.array(X/255.0)
y = np.array(y)
```

```python
checkpoint_filepath = os.path.join(os.getcwd(),'/tmp/checkpoint')

def create_model(CHECKPOINT_PATH):
  model = Sequential()

  model.add(Conv2D(256, (4,4), input_shape=X.shape[1:], activation='relu'))
  model.add(MaxPooling2D(pool_size=(3,3)))
  model.add(Dropout(0.2))

  model.add(Conv2D(156, (3,3), activation='relu'))
  model.add(AveragePooling2D(pool_size=(2,2)))
  model.add(Dropout(0.2))

  model.add(Flatten())

  model.add(Dense(100, activation='relu'))
  model.add(Dense(50, activation='relu'))
  model.add(Dense(1, activation='sigmoid'))

  opt = keras.optimizers.Adam(lr=0.00009)
  model.compile(
      loss='binary_crossentropy',
      optimizer='adam',#opt,
      metrics=['accuracy']
  )

  cp_callback = keras.callbacks.ModelCheckpoint(CHECKPOINT_PATH,
                                                 save_best_only=True,
                                                 verbose=1,
                                                 monitor='val_loss'
                                                 )
  
  return model, cp_callback

model,cp_callback = create_model(checkpoint_filepath)
```

```python
model.load_weights('model.h5')
```

```python
def pred(img):
    img_array = cv2.resize(cv2.imread(os.path.join(os.path.join(os.getcwd(),'static/temp'), img), cv2.IMREAD_COLOR), (IMG_SIZE, IMG_SIZE))
    img_array = np.array(img_array).reshape(-1, IMG_SIZE, IMG_SIZE, 3)
    return model.predict(img_array)
```
